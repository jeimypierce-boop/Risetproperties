```python
import pathlib

root = pathlib.Path(r"c:\xampp\htdocs\RisetProperties-Website-and-AdminPanel")
changed = []
for path in root.rglob("*.php"):
    if path.name == "fix_settings_links.ipynb":
        continue
    text = path.read_text(encoding="utf-8")
    new_text = text.replace('href="admin-panel-setting.html"', 'href="admin-system-settings.php?section=overview"').replace('href="admin-setting.html"', 'href="admin-system-settings.php?section=overview"')
    if new_text != text:
        path.write_text(new_text, encoding="utf-8")
        changed.append(str(path))
for path in root.rglob("*.html"):
    text = path.read_text(encoding="utf-8")
    new_text = text.replace('href="admin-panel-setting.html"', 'href="admin-system-settings.php?section=overview"').replace('href="admin-setting.html"', 'href="admin-system-settings.php?section=overview"')
    if new_text != text:
        path.write_text(new_text, encoding="utf-8")
        changed.append(str(path))
print(f"Updated {len(changed)} files")
for p in changed[:50]:
    print(p)
```
